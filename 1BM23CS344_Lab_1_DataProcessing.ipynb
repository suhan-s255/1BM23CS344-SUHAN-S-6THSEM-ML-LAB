```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

diabetes = pd.read_csv("diabetes.csv")
adult = pd.read_csv("adult.csv")

diabetes.fillna(diabetes.median(), inplace=True)

q1 = diabetes.quantile(0.25)
q3 = diabetes.quantile(0.75)
iqr = q3 - q1
diabetes = diabetes[~((diabetes < (q1 - 1.5 * iqr)) | (diabetes > (q3 + 1.5 * iqr))).any(axis=1)]

minmax = MinMaxScaler()
std = StandardScaler()

diabetes_scaled_minmax = pd.DataFrame(minmax.fit_transform(diabetes), columns=diabetes.columns)
diabetes_scaled_std = pd.DataFrame(std.fit_transform(diabetes), columns=diabetes.columns)

adult.replace("?", np.nan, inplace=True)
adult.fillna(adult.mode().iloc[0], inplace=True)

cat_cols = adult.select_dtypes(include="object").columns
adult = pd.get_dummies(adult, columns=cat_cols, drop_first=True)

q1 = adult.quantile(0.25)
q3 = adult.quantile(0.75)
iqr = q3 - q1
adult = adult[~((adult < (q1 - 1.5 * iqr)) | (adult > (q3 + 1.5 * iqr))).any(axis=1)]

adult_scaled_minmax = pd.DataFrame(minmax.fit_transform(adult), columns=adult.columns)
adult_scaled_std = pd.DataFrame(std.fit_transform(adult), columns=adult.columns)
```

1. Missing Values

Diabetes Dataset:
Some numerical columns contained missing values. These were handled by replacing missing values with the median of each column to avoid the effect of outliers.

Adult Income Dataset:
Missing values were represented using "?" in categorical columns such as workclass, occupation, and native-country. These were replaced with NaN and then filled using the mode (most frequent value) of each column.

2. Categorical Columns and Encoding

Diabetes Dataset:
The diabetes dataset contains only numerical features, so no categorical encoding was required.

Adult Income Dataset:
Categorical columns identified include workclass, education, marital-status, occupation, relationship, race, sex, and native-country.
These were encoded using One-Hot Encoding (get_dummies) to convert categorical values into numerical form suitable for machine learning models.

3. Difference Between Min-Max Scaling and Standardization

Min-Max Scaling:
Transforms data into a fixed range, usually between 0 and 1. It is sensitive to outliers and is useful when the data does not follow a normal distribution.

Standardization (Standard Scaler):
Centers the data around a mean of 0 with a standard deviation of 1. It works well when the data follows a normal distribution and is less affected by outliers.

Usage:
Min-Max Scaling is preferred for algorithms like KNN and neural networks, while Standardization is commonly used for algorithms like linear regression, logistic regression, and SVM.
